### **Task 06**: Add a New Column 

```python
df["Age_Group"] = 1
```

```python
def categorize_age(age):
    if age < 18:
        return "Minor"
    elif age <= 25:
        return "Young Adult"
    elif age <= 40:
        return "Adult"
    else:
        return "Older Adult"

df['Age_Group'] = df['Age'].apply(categorize_age)
df.to_csv('Daily_Arrests_Updated.csv', index=False)
```

```python
df.head(10)
```

```python
df.tail(10)
```

#### This new column groups individuals into age categories (Minor, Young Adult, Adult, and Older Adult), which makes it easier to analyze arrest patterns and observe the relationship between age and criminal activity. Additonally, age grouping aligns more closely with how policies and interventions are designed. For example, programs for minors, young adults, or older adults differ, so grouping ages makes the data more useful for real-world decision-making.

### **Task 01**: Individual Visualizations

```python
import pandas as pd 
import matplotlib.pyplot as plt 

df = pd.read_csv("Daily_Arrests.csv")
df.columns = df.columns.str.strip()
```

```python
plt.figure()
df['AGE'].hist(bins=20, color='lavender', edgecolor='purple')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Distribution of Age")
```

#### This histogram helps us visualize the frequency of different age values. The lines divide the age data into 20 bins and counts how many individuals fall into each bin.

```python
plt.figure()
df['OFFENSE'].value_counts().head(10).plot(
    kind='bar',colormap='viridis')
plt.xlabel("Offense")
plt.ylabel("Count") 
plt.title("Top 10 Most Common Offenses")
plt.xticks(rotation=90)
plt.show()
```

#### The bar chart is used to visualize the most common offense. Through this visualization, it is illustrated that Assault in the second degree is the most common offense committed. This visualization can lead to better targeted interventions to decrease such occurances 

```python
plt.figure()
df['CITY'].value_counts().head(10).plot(kind='bar', color='lavender')
plt.xlabel("City")
plt.ylabel("Number of Arrests")
plt.title("Top Cities by Arrest Count")
plt.xticks(rotation=90)
plt.show()
```

#### The bar graph is used to display the number of arrests that occur in each city. This visualization could provide insight into crime hotspots and geographical patterns. The box chart shows that Silver Spring and Rockville are major crime hotspots. 

```python
import seaborn as sns 
plt.figure()
plt.boxplot(df['AGE'])
plt.ylabel("Age")
plt.title("Boxplot of Age Distribution")
plt.show()
```

#### The boxplot is used to summarize and visualize the distribution of age, including its central tendency, spread, and  outliers.As we can see from the box plot, the majority of arrests were made within the age range of 25 to 45. However, there are a few outliers as we see some perpatrators are above the age of 65.

### **Task 02**: Comparative Visualizations

```python
df['OFFENSE'].nunique()
```

```python
counts = df['OFFENSE'].value_counts()
counts[counts > 5]
```

```python
frequent_offenses = counts[counts > 10].index
filtered_df = df[df['OFFENSE'].isin(frequent_offenses)]
```

#### The visualization shows the average ages of perpetrators across the most common offense type. This visualization suggests that different crimes are associated with different age groups. For example, offenses such as armed robbery may involve younger individuals, while other offenses, such as second-degree arson, show higher average ages.

```python
plt.figure(figsize=(10,6))
df.groupby('CITY')['AGE'].mean().head(30).plot(kind='bar')
plt.xlabel("City")
plt.ylabel("Age")
plt.title("Relationship Between Average Age and City")
plt.show()
```

### **Task 03**: Relationship Analysis 

```python
import matplotlib.pyplot as plt
df['ARREST DATE'] = pd.to_datetime(df['ARREST DATE'])
df['Month'] = df['ARREST DATE'].dt.month
df['Arrest_Day'] = df['ARREST DATE'].dt.day
march_data = df[df['Month'] == 3].copy()
plt.figure(figsize=(16, 10))
sns.scatterplot(data=df, x="ARREST DATE", y="AGE", color="Green")
plt.title("Age vs Days in March")
plt.xlabel("Days in March")
plt.ylabel("Age")
```

```python
heatmap_data = pd.crosstab(march_data["AGE"], march_data["Arrest_Day"])
march_corr = heatmap_data.corr()

plt.figure(figsize=(16, 10))
sns.heatmap(march_corr, annot=False, cmap="spring", cbar_kws={"ticks": [1, 2, 3, 4, 5], "label": "Number of Arrests"})
plt.title("Age vs Days in March")
plt.xlabel("Days in March")
plt.ylabel("Age")
```

#### Since the data consisted of arrests from different people in different states, it was expected that the scatter plot would likely not show a correlation between the two variables. As predicted, the scatter plot of Age vs Days in March yielded no correlation between the two variables. In the heatmap, given that there are a select few boxes that are very red, it shows that many of the people included in the data had more than one offense. Interestingly, the red boxes are lined up in a diagonal line, mimicking a negative slope line, showing that the older the people were, the more likley their only offense occured later in the month of March.
