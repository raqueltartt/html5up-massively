###### a tuple is immutabile, so you cannot replace a variable. Once something in a tuple is defined, it cannot be altered.

```python
student_info = ("4002111", "Raquel Tartt", "4.0")
```

```python
student_id, student_name, student_gpa = student_info
```

```python
print("Student ID:", student_id)
print("Name:", student_name)
print("GPA:", student_gpa)
```

###### I would use a tuple in a data science project if the analyzed variables were not subject to change. To protect defined variables from being rewritten, tuples can secure them as they are immutable. For example, If I wanted to analyze student final grades for a semester, I could use a tuple, since none of the data will be subject to change.
