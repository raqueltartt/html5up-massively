### **Task 06**: Add a New Column 

```python
df["Age_Group"] = 1
```

```python
def categorize_age(age):
    if age < 18:
        return "Minor"
    elif age <= 25:
        return "Young Adult"
    elif age <= 40:
        return "Adult"
    else:
        return "Older Adult"

df['Age_Group'] = df['Age'].apply(categorize_age)
df.to_csv('Daily_Arrests_Updated.csv', index=False)
```

```python
df.head(10)
```

```python
df.tail(10)
```

#### This new column groups individuals into age categories (Minor, Young Adult, Adult, and Older Adult), which makes it easier to analyze arrest patterns and observe the relationship between age and criminal activity. Additonally, age grouping aligns more closely with how policies and interventions are designed. For example, programs for minors, young adults, or older adults differ, so grouping ages makes the data more useful for real-world decision-making.

### **Task 09**: Initial Observations

#### Exploring the Montgomery County Daily Arrests dataset revealed several compelling patterns and raised important questions. One of the most striking findings was the frequent appearance of “NO FIXED ADDRESS” as a listed street, suggesting that a notable portion of individuals being arrested may be experiencing housing instability, especially in Rockville, MD. Theft emerged as one of the most common offense types across all age groups, particularly among individuals who noted having no fixed address, indicating that many arrests may be economically motivated rather than violent in nature. Notably, older adults (45+) appear disproportionately in theft-related arrests concentrated in Rockville, highlighting how age-related factors such as limited income, unemployment, or reduced social support may influence criminal behavior. Additionally, theft and assault emerged as the most frequent offenses in minor and young adult populations, which can meaningfully drive targeted reforms, school-based prevention programs, and youth engagement strategies.

#### The dataset also contained missing values, primarily in the Middle Name column, which, while not critical to analysis, reflects inconsistencies in data entry practices. Additionally, the wide variation in offense descriptions, such as multiple tiers of theft based on dollar amounts, introduced challenges in categorization and required careful interpretation during analysis.

#### Overall, these patterns suggest that arrests in Montgomery County may be shaped not only by individual behavior but also by broader social and economic conditions. Furthermore, the geographic variation across cities like Rockville, Silver Spring, and Laurel raises important questions about whether differences in arrest patterns reflect underlying community conditions or variations in policing practices.
